```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time

chromedriver_path = 'C:\\Users\\vicky\\OneDrive\\Desktop\\chromedriver-win64\\chromedriver.exe'

try:
    service = Service(chromedriver_path)
    driver = webdriver.Chrome(service=service)

    driver.get('https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile')
    time.sleep(5)
    driver.maximize_window()

    wait = WebDriverWait(driver, 20)

    LINKS_KERALA = []
    ROUTE_KERALA = []

    def Kerala_link_route(path):
        try:
            routes = driver.find_elements(By.XPATH, path)
            for route in routes:
                route_link = route.get_attribute("href")
                route_name = route.text
                LINKS_KERALA.append(route_link)
                ROUTE_KERALA.append(route_name)

            for page_number in range(1, 3):
                try:
                    pagination_container = wait.until(
                        EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]'))
                    )

                    next_button = pagination_container.find_element(
                        By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]'
                    )

                    actions = ActionChains(driver)
                    actions.move_to_element(next_button).perform()
                    time.sleep(2)

                    next_button.click()

                    wait.until(
                        EC.text_to_be_present_in_element(
                            (By.XPATH, '//*[@class="DC_117_pageTabs DC_117_pageActive"]'), str(page_number + 1)
                        )
                    )

                    routes = driver.find_elements(By.XPATH, path)
                    for route in routes:
                        route_link = route.get_attribute("href")
                        route_name = route.text
                        LINKS_KERALA.append(route_link)
                        ROUTE_KERALA.append(route_name)

                except Exception:
                    break

        except Exception:
            pass

        return LINKS_KERALA, ROUTE_KERALA

    LINKS_KERALA, ROUTE_KERALA = Kerala_link_route("//a[@class='route']")

    print(f"Total routes scraped: {len(ROUTE_KERALA)}")
    for i in range(min(5, len(ROUTE_KERALA))):
        print(f"Route: {ROUTE_KERALA[i]} - Link: {LINKS_KERALA[i]}")

finally:
    driver.quit()
```

Andhra routes

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time

# Set up chromedriver path and initialize driver
chromedriver_path = 'C:\\Users\\vicky\\OneDrive\\Desktop\\chromedriver-win64\\chromedriver.exe'

try:
    service = Service(chromedriver_path)
    driver = webdriver.Chrome(service=service)

    # Navigate to the RedBus Andhra routes page
    driver.get('https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile')
    time.sleep(5)
    driver.maximize_window()

    # Initialize WebDriverWait
    wait = WebDriverWait(driver, 20)

    # Andhra routes data
    LINKS_ANDHRA = []
    ROUTE_ANDHRA = []

    def Andhra_link_route(path):
        try:
            # Extract routes from the current page
            routes = driver.find_elements(By.XPATH, path)
            for route in routes:
                route_link = route.get_attribute("href")
                route_name = route.text
                LINKS_ANDHRA.append(route_link)
                ROUTE_ANDHRA.append(route_name)

            print(f"Collected {len(ROUTE_ANDHRA)} routes on the first page")

            # Pagination navigation
            for page_number in range(1, 6):  # Change range based on the total number of pages
                try:
                    # Locate the pagination container
                    pagination_container = wait.until(
                        EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]'))
                    )

                    # Identify the "Next Page" button
                    next_button = pagination_container.find_element(
                        By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]'
                    )

                    # Use ActionChains to interact with the button
                    actions = ActionChains(driver)
                    actions.move_to_element(next_button).perform()
                    time.sleep(2)

                    # Click the "Next Page" button
                    next_button.click()

                    # Wait for the next page to load
                    wait.until(
                        EC.text_to_be_present_in_element(
                            (By.XPATH, '//*[@class="DC_117_pageTabs DC_117_pageActive"]'), str(page_number + 1)
                        )
                    )

                    # Collect routes from the new page
                    routes = driver.find_elements(By.XPATH, path)
                    for route in routes:
                        route_link = route.get_attribute("href")
                        route_name = route.text
                        LINKS_ANDHRA.append(route_link)
                        ROUTE_ANDHRA.append(route_name)

                    print(f"Collected {len(ROUTE_ANDHRA)} routes after moving to page {page_number + 1}")

                except Exception as e:
                    print(f"No more pages or error: {e}")
                    break

        except Exception as e:
            print(f"Error collecting data: {e}")

        return LINKS_ANDHRA, ROUTE_ANDHRA

    # Extract Andhra routes
    LINKS_ANDHRA, ROUTE_ANDHRA = Andhra_link_route("//a[@class='route']")

    # Print summary
    print(f"Total routes scraped: {len(ROUTE_ANDHRA)}")
    for i in range(min(5, len(ROUTE_ANDHRA))):
        print(f"Route: {ROUTE_ANDHRA[i]} - Link: {LINKS_ANDHRA[i]}")

finally:
    # Close the WebDriver
    driver.quit()
```

```python
df_andhra = pd.DataFrame({"Route_name": ROUTE_ANDHRA, "Route_link": LINKS_ANDHRA})
print(df_andhra)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time

chromedriver_path = 'C:\\Users\\vicky\\OneDrive\\Desktop\\chromedriver-win64\\chromedriver.exe'

try:
    service = Service(chromedriver_path)
    driver = webdriver.Chrome(service=service)

    driver.get('https://www.redbus.in/online-booking/astc/?utm_source=rtchometile')
    time.sleep(5)
    driver.maximize_window()

    wait = WebDriverWait(driver, 20)

    LINKS_ASSAM = []
    ROUTE_ASSAM = []

    def Assam_link_route(path):
        try:
            routes = driver.find_elements(By.XPATH, path)
            for route in routes:
                route_link = route.get_attribute("href")
                route_name = route.text
                LINKS_ASSAM.append(route_link)
                ROUTE_ASSAM.append(route_name)

            for page_number in range(1, 6):
                try:
                    pagination_container = wait.until(
                        EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]'))
                    )

                    next_button = pagination_container.find_element(
                        By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]'
                    )

                    actions = ActionChains(driver)
                    actions.move_to_element(next_button).perform()
                    time.sleep(2)

                    next_button.click()

                    wait.until(
                        EC.text_to_be_present_in_element(
                            (By.XPATH, '//*[@class="DC_117_pageTabs DC_117_pageActive"]'), str(page_number + 1)
                        )
                    )

                    routes = driver.find_elements(By.XPATH, path)
                    for route in routes:
                        route_link = route.get_attribute("href")
                        route_name = route.text
                        LINKS_ASSAM.append(route_link)
                        ROUTE_ASSAM.append(route_name)

                except Exception:
                    break

        except Exception:
            pass

        return LINKS_ASSAM, ROUTE_ASSAM

    LINKS_ASSAM, ROUTE_ASSAM = Assam_link_route("//a[@class='route']")

    print(f"Total routes scraped: {len(ROUTE_ASSAM)}")
    for i in range(min(5, len(ROUTE_ASSAM))):
        print(f"Route: {ROUTE_ASSAM[i]} - Link: {LINKS_ASSAM[i]}")

finally:
    driver.quit()
```
